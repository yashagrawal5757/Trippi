```python
import pandas as pd
from google.colab import files
```

```python
la_reviews_df  = pd.read_csv("/content/la_reviews.csv")
nyc_reviews_df  = pd.read_csv("/content/nyc_reviews.csv")
austin_reviews_df  = pd.read_csv("/content/austin_reviews.csv")
chicago_reviews_df  = pd.read_csv("/content/chicago_reviews.csv")
```

```python
concatenated_df = pd.concat([la_reviews_df, nyc_reviews_df, austin_reviews_df, chicago_reviews_df], axis=0)
```

```python
concatenated_df
```

```python
concatenated_df.columns
```

```python
print(concatenated_df.isnull().sum())
```

```python
concatenated_df.drop(concatenated_df[concatenated_df['comments'].isnull()].index, inplace=True)
```

```python
concatenated_df.drop(concatenated_df[concatenated_df['reviewer_name'].isnull()].index, inplace=True)
```

```python
print(concatenated_df.isnull().sum())
```

```python
print(concatenated_df.dtypes)
```

```python
concatenated_df['date'] = pd.to_datetime(concatenated_df['date'])
```

```python
print(concatenated_df.dtypes)
```

```python
concatenated_df.rename(columns={'id': 'rev_id'}, inplace=True)
```

```python
concatenated_df
```

```python
duplicates = concatenated_df[concatenated_df.duplicated(['rev_id'], keep=False)]
duplicates
```

```python
concatenated_df = concatenated_df.drop_duplicates(subset=['rev_id'])
reviews_df  = concatenated_df
reviews_df
```
